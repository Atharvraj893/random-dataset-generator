```python
import numpy as np
import pandas as pd
import random as rd
from faker import Faker
```

```python
print("What type of dataset you want :\n1. Student marks\n2. Employee salary\n3. Sales dataset")
dt = int(input("Enter your choice from 1/2/3 = "))
fake = Faker("en_IN")
if dt==1:
    di = {}
    print("You get Student marks data with study hour and marks of 5 subjects and name of student")
    row = int(input("Enter how many rows you want : "))
    for i in range(row):
        sh = rd.randint(1,10)
        Chemistry = (sh*10)+rd.randint(-5,5)
        Physics = (sh*10)+rd.randint(-5,5)
        Computer = (sh*10)+rd.randint(-5,5)
        English = (sh*10)+rd.randint(-5,5)
        Maths = (sh*10)+rd.randint(-5,5)
        value = [sh,Physics,Chemistry,Computer,English,Maths]
        n = fake.name()
        di[n] = value
    df = pd.DataFrame.from_dict(di,orient="index",columns = ["Study Hour","Physics","Chemistry","Computer","English","Maths"])
    df = df.reset_index()
    df.rename(columns={"index":"Name"},inplace=True)
    print(df)
elif dt==2:
    ei = {}
    print("You get employee salary dataset with employee id name city department etc....")
    row = int(input("Enter how many rows you want : "))
    ids = []
    for i in range(row):
        city = fake.city()
        l = ["IT","HR","Mareketing","Support"]
        dep = rd.choice(l)
        n = fake.name()
        salary = rd.randint(30000,100000)
        value = [n,dep,salary,city]
        new_id = "STU-"+fake.bothify(text="??!!##")
        if new_id not in ids:
            ids.append(new_id)
            ei[new_id] = value
    df1 = pd.DataFrame.from_dict(ei,orient="index",columns = ["Name","Department","Salary","City"])
    df1 = df1.reset_index()
    df1.rename(columns={"index":"Id"},inplace=True)
    print(df1)
```
